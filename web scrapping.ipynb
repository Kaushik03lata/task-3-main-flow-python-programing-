```python
import requests as rst
from bs4 import BeautifulSoup
```

```python
link='https://www.geeksforgeeks.org/what-is-web-scraping-and-how-to-use-it/'
```

```python
respond=rst.get(link)
```

```python
if respond.status_code==200:
    bsp=BeautifulSoup(respond.text,'html.parser')
```

```python
page_text=bsp.get_text()
page_text
```

```python
limits=[a['href']for a in bsp.find_all('a',href=True)]
limits
```

```python
images=[img['src']for img in bsp.find_all('img',src=True)]
images
```
